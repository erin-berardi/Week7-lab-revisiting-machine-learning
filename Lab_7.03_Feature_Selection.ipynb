```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
# import warnings
# warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import VarianceThreshold 
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import r2_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
```

```python
numerical = pd.read_csv('numerical_target.csv')
categorical = pd.read_csv('categorical.csv')
targets = numerical[['TARGET_B','TARGET_D']]
```

```python
numerical.drop(['Unnamed: 0','TARGET_B','TARGET_D'],axis=1,inplace = True)
categorical.drop('Unnamed: 0',axis=1, inplace = True)
```

```python
X=pd.concat([numerical,categorical], axis=1)
X.drop(['SOLIH','VETERANS','MDMAUD'], axis=1, inplace = True)
```

```python
X.isna().sum().sum()
```

```python
y=targets['TARGET_B']
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)
print('X_train shape is:',X_train.shape)
print('y_train shape is:', y_train.shape)
print('X_test shape is:', X_test.shape)
print('y_train shape is:', y_test.shape)
```

# Encoding and Scaling train and test

```python
X_train_cat= X_train.select_dtypes(include=object)
X_train_num= X_train.select_dtypes(include=np.number)
X_test_cat= X_test.select_dtypes(include=object)
X_test_num= X_test.select_dtypes(include=np.number)
```

```python
encoder = OneHotEncoder(drop='first').fit(X_train_cat)

cols = encoder.get_feature_names(input_features=X_train_cat.columns)

X_train_cat_encode = pd.DataFrame(encoder.transform(X_train_cat).toarray(),columns=cols)

X_train_cat_encode.head()
```

```python

cols = encoder.get_feature_names(input_features=X_test_cat.columns)

X_test_cat_encode = pd.DataFrame(encoder.transform(X_test_cat).toarray(),columns=cols)

X_test_cat_encode.head()
```

```python
transformer = MinMaxScaler().fit(X_train_num)
X_train_norm = transformer.transform(X_train_num)
X_train_num_scale = pd.DataFrame(X_train_norm, index = X_train_num.index, columns=X_train_num.columns).reset_index(drop=True)
X_train_num_scale.head()
```

```python
X_test_norm = transformer.transform(X_test_num)
X_test_num_scale = pd.DataFrame(X_test_norm, index = X_test_num.index, columns=X_test_num.columns).reset_index(drop=True)
X_test_num_scale.head()
```

```python
X_train = pd.concat([X_train_num_scale,X_train_cat_encode], axis = 1)#.reset_index(drop=True)
X_train
```

```python
X_test = pd.concat([X_test_num_scale,X_test_cat_encode], axis = 1)#.reset_index(drop=True)
X_test
```

```python

clf = RandomForestClassifier(max_depth=5,
                             min_samples_split=20,
                             min_samples_leaf =20)
clf.fit(X_train, y_train)
print('TrainSet = ',clf.score(X_train, y_train))
print('TestSet = ',clf.score(X_test, y_test))
```

# Looking at Variance

```python
numerical
```

```python
transformer = MinMaxScaler().fit(numerical)
num_scaled = transformer.transform(numerical)
num_scale = pd.DataFrame(num_scaled, index = numerical.index, columns=numerical.columns).reset_index(drop=True)
num_scale.reset_index(drop=True)
num_scale.head()
```

```python

var_threshold = 0.02
sel = VarianceThreshold(threshold=(var_threshold))

# 1) default is threshold = 0 :eliminate all coumns that are identical for all rows
# 2) in practice we would scale the columns first, and then apply threshold, or apply 
#    a different threshold for different columns

# This drops the columns that have a variance less than this threshold
sel = sel.fit(num_scale)
temp = sel.transform(num_scale)
temp = pd.DataFrame(temp)
print(num_scale.shape)
print(temp.shape)
```

```python
sel.variances_
```

```python
sel.variances_ > var_threshold
sel.get_support()
var_list = list(sel.get_support())
var_list
```

```python
list(zip(numerical.columns, var_list))
```

```python
variance_to_drop = [col[0] for col in zip(numerical.columns, var_list) if col[1] == False]
variance_to_drop
```

```python
# So these are the 240 columns we can drop
# if we also want to see the actual variances we can use this:
```

```python
removed_columns = pd.DataFrame(data=(numerical.columns,sel.variances_,sel.get_support()), index=('column_name','variance','statement')).T
removed_columns.loc[(removed_columns['statement'] == False),:]
```

```python
num_dropped_variance = numerical.drop(columns = variance_to_drop)
num_dropped_variance 
```
